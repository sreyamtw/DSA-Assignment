**4. Fill in all the missing values present in all the columns in the dataset.**

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data['Age'].isna().sum()
```

**5. Check and handle outliers in at least 3 columns in the dataset.**

```python
sns.boxplot(y=data['Age'])
plt.show()
```

```python
Q1_age=np.percentile(data['Age'],25)
Q3_age=np.percentile(data['Age'],75)
IQR_age=Q3_age-Q1_age
l_age=Q1_age-1.5*IQR_age
u_age=Q3_age+1.5*IQR_age
outlier_age=[]
for x in data['Age']:
  if x>u_age or x<l_age:
    outlier_age.append(x)
```

```python
len(outlier_age)
```

```python
ind_age =(data['Age']>u_age )| (data['Age']<l_age)
location =data.loc[ind_age].index
location
```

```python
data.drop(location,inplace=True)
```
