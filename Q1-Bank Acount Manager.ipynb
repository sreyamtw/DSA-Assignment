```python
def display_balance():
    print('\nYour balance is Rs.',balance)

def deposit():
    amount = float(input('Enter the amount to be deposited:'))

    if amount <= 0:
        print('\nInvalid amount!')
        return 0
    else:
        print('\nThe amount has been deposited')
        return amount

def withdraw():
    amount = float(input('Enter the amount to be withdrawn:'))

    if amount > balance:
        print('\nInsufficient funds!')
        return 0
    elif amount <= 0:
        print('Amount must be greater than 0')
        return 0
    else:
        print('\nThe Rs.',amount,'has been withdrawn')
        return amount


balance = 0.0
flag=True

while flag:
    print('\n-----Choose the operation-----')
    print('\n1.Balance Inquiry \n2.Deposit \n3.Withdraw \n4.Exit')

    choice = input('Enter the operation no.: ')

    if choice == '1':
        display_balance()
    elif choice == '2':
        balance += deposit()
        display_balance()
    elif choice == '3':
        balance -= withdraw()
    elif choice == '4':
        flag = False
    else:
        print("\nInvalid choice!")

```
